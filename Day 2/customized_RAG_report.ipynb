```python
from google.colab import files

uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]
```

Step 3: Load and Split PDF Content

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader(pdf_path)
docs = loader.load_and_split()

print(f"Loaded {len(docs)} document chunks.")
```

Step 4: Embed Documents and Build FAISS Vector Store

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

# Use a local embedding model
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Create FAISS vector store
vectorstore = FAISS.from_documents(docs, embeddings)

# Save locally for faster retrieval later
vectorstore.save_local("financial_report_index")
```

Load LLM and Create Retrieval QA Chain

```python
from langchain.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# Load tokenizer and model
model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Create HF pipeline
pipe = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=200,
    device=0 if torch.cuda.is_available() else -1
)

llm = HuggingFacePipeline(pipeline=pipe)

# Load the FAISS index
vectorstore = FAISS.load_local(
    "financial_report_index",
    embeddings,
    allow_dangerous_deserialization=True
)

# Create Retrieval QA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vectorstore.as_retriever()
)
```

Step 6: Ask Questions & Generate Insights

```python
# Reduce context by limiting retrieved documents
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vectorstore.as_retriever(max_docs=2),
    return_source_documents=False
)

# Ask question
query = "What is the company's total revenue for the last fiscal year?"
response = qa_chain.invoke({"query": query})  # Using invoke instead of run

print("Answer:", response['result'])
```

```python
queries = [
    "What is the net income?",
    "Who are the major stakeholders?",
    "Summarize the risks mentioned in the report.",
    "What is the dividend policy?"
]

for q in queries:
    res = qa_chain.invoke({"query": q})
    print(f"\nQ: {q}\nA: {res['result']}")
```

Step 7: Generate a Custom RAG Report

```python
def generate_rag_report(questions):
    report = {}
    for q in questions:
        print(f"Processing: {q}")
        ans = qa_chain.run(q)
        report[q] = ans
    return report

# Define your report template with relevant questions
report_questions = [
    "What is the company's total revenue?",
    "What was the net income for the latest year?",
    "Who are the major competitors?",
    "What are the strategic goals for next year?",
    "Summarize the financial health of the company."
]

rag_report = generate_rag_report(report_questions)

# Print the final RAG-powered report
print("\n\n📊 GENERATED RAG REPORT:")
for q, a in rag_report.items():
    print(f"\nQ: {q}\nA: {a}")
```

Download the Report as a Text File

```python
from google.colab import files

with open('financial_rag_report.txt', 'w') as f:
    f.write("📊 GENERATED RAG FINANCIAL REPORT\n")
    f.write("="*40 + "\n\n")
    for q, a in rag_report.items():
        f.write(f"Q: {q}\nA: {a}\n\n" + "-"*40 + "\n\n")

files.download('financial_rag_report.txt')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import base64
from io import BytesIO

def generate_chart_base64(df):
    if df.empty:
        return ""

    try:
        plt.figure(figsize=(10, 5))

        if len(df.columns) == 1:
            sns.countplot(y=df.iloc[:, 0])
            plt.title("Frequency Distribution")
        else:
            sns.barplot(data=df, x=df.columns[0], y=df.columns[1])
            plt.xticks(rotation=45)
            plt.title(f"{df.columns[1]} vs {df.columns[0]}")

        plt.tight_layout()

        buf = BytesIO()
        plt.savefig(buf, format='png', dpi=100)
        plt.close()
        buf.seek(0)
        return base64.b64encode(buf.getvalue()).decode('utf-8')

    except Exception as e:
        print(f"Chart generation failed: {str(e)}")
        return ""
```
